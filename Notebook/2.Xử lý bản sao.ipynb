```python
# Matplotlib l mt th vin v  th cho ngn ng lp trnh Python
# v phn m rng ton hc s NumPy ca n
from matplotlib import pyplot as plt
import cv2 
image = cv2.imread('image\icon.png', cv2.IMREAD_COLOR)
img_1 = image.copy()
(height, width, channel) = img_1.shape
for i in range(height//2):
    for j in range(width//2):
        img_1[i, j] = (200, 100, 255)

# Chuyn knh mu cho ph hp vi th vin matplotlib
(b,g,r) = cv2.split(img_1)
img_2 = cv2.merge([r,g,b])
plt.imshow(img_2)
plt.title('my picture')
plt.show()
```
