#### _Bài tập 1:_ Nhận xét sự khác nhau khi kết hợp các kênh màu thông theo thứ tự. (hiệu suất nhận diện ảnh có thể được nâng cao nhờ việc thay đổi thứ tự kênh màu)

```python
# Matplotlib là một thư viện vẽ đồ thị cho ngôn ngữ lập trình Python
# và phần mở rộng toán học số NumPy của nó
from matplotlib import pyplot as plt
import cv2 

def Show(image, name):
    # Chuyển kênh màu cho phù hợp với thư viện matplotlib
    (b, g, r) = cv2.split(image)
    img_2 = cv2.merge([r,g,b])
    plt.imshow(img_2)
    plt.title(name)

def ShowChannel(b, g, r, name):
    img = cv2.merge([b,g,r])
    Show(img, name)
    
image = cv2.imread(r'd:\icon.png', cv2.IMREAD_COLOR)
img_1 = image.copy()
(b, g, r) = cv2.split(image)
# Hình vẽ có 1 hàng 2 cột, vẽ ở vị trí thứ 1
plt.subplot(231), ShowChannel(b, g, r, "Ghép kênh b, g, r")
plt.subplot(232),ShowChannel(b, r, g, "Ghép kênh b, r, g")
plt.subplot(233),ShowChannel(r, g, b, "Ghép kênh r, g, b")
plt.subplot(234),ShowChannel(r, b, g, "Ghép kênh r, b, g")
plt.subplot(235),ShowChannel(g, b, r, "Ghép kênh g, b, r")
plt.subplot(236),ShowChannel(g, r, b, "Ghép kênh g, r, b")
plt.show()
```
