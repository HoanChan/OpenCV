```python
from matplotlib import pyplot as plt
import cv2 

def Show(image, name):
    # Chuyn knh mu cho ph hp vi th vin matplotlib
    (b, g, r) = cv2.split(image)
    img_2 = cv2.merge([r,g,b])
    plt.imshow(img_2)
    plt.title(name)
    plt.show()

def ShowChannel(b, g, r, name):
    img = cv2.merge([b,g,r])
    Show(img, name)
    
image = cv2.imread(r'd:\icon.png', cv2.IMREAD_COLOR)
img_1 = image.copy()
(w, h, c) = image.shape
image_scaling_up = cv2.resize(image, (w*2,h*2),interpolation=cv2.INTER_LINEAR)
image_scaling_down = cv2.resize(image, (w//2,h//2),interpolation=cv2.INTER_LINEAR)
Show(image, "x1")
Show(image_scaling_up, "x4")
Show(image_scaling_down, "1/4")
```
